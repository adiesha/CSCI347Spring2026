```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
dp_df = pd.read_csv('data_points.csv')
dp_df.head()
D = dp_df.to_numpy()
```

```python
sigma = np.cov(D.T)
print(sigma)
```

```python
np.trace(sigma)
```

```python
Z = D-D.mean(axis=0)
standard_scalar = StandardScaler(with_std=False)
Z_2 = standard_scalar.fit_transform(D)
```

```python
np.cov(Z_2.T)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit_transform(Z_2)
pca.components_
```

```python
pca.explained_variance_ratio_
```
